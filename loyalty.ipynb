```python
#импортирую библитеку

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#импортирую данные

df=pd.read_csv('https://stepik.org/media/attachments/lesson/383523/lesson_3_data__1_.csv', encoding='windows-1251')
```

```python
df
```

```python
#переименовываю колонки

df=df.rename(columns={'tc':'user_id', 'art_sp': 'brand_info'})
```

```python
#"достаю" имя бренда

df['brand_name']=df.brand_info.apply(lambda x: x.split(' ')[-1])
```

```python
df
```

```python
#нахожу кол-во покупок для каждого пользователя и выбираю тех пользователей, у которых кол-во покупок содержится в верхнем 25%-ном квантиле

users_purchases = df.groupby('user_id', as_index=False) \
    .agg({'brand_info': 'count'}) \
    .rename(columns={'brand_info': 'purchases'}) \
    .query('purchases >=purchases.quantile(q=0.75)')
```

```python
users_purchases
```

```python
#определяю любимый бренд у каждого покупателя как часто покупаемый и кол-во покупок любимого бренда


favourite_brand_purchases_df =df.groupby(['user_id', 'brand_name'], as_index=False) \
    .agg({'brand_info':'count'}) \
    .sort_values(['user_id', 'brand_info'], ascending= [False, False]) \
    .groupby('user_id') \
    .head(1) \
    .rename(columns=({'brand_name': 'favourite_brand', 'brand_info': 'favourite_brand_purchases'}))
```

```python
favourite_brand_purchases_df
```

```python
#определяю кол-во брендов, покупаемых клиентом

user_unique_brands = df.groupby('user_id', as_index=False) \
    .agg({'brand_name': pd.Series.nunique}) \
    .rename(columns={'brand_name': 'unique_brands'})
```

```python
user_unique_brands
```

```python
#соединяю все 3 таблицы

loyalty_df=users_purchases.merge(user_unique_brands, on='user_id') \
    .merge(favourite_brand_purchases_df, on ='user_id')
```

```python
loyalty_df.head()
```

```python
#есть ли прямая зависимость между кол-вом брендов и покупками

sns.lmplot(data=loyalty_df, x = 'unique_brands', y = 'purchases')
```

```python
#1 вариант: "супер" лояльные клиенты - те, которые покупают только 1 бренд

loyal_users=loyalty_df[loyalty_df.unique_brands==1]
```

```python
#Визуализирую кол-во "супер" лояльных клиентов у каждого бренда

data_pic1=loyal_users.groupby('favourite_brand', as_index=False).agg({'user_id':'count'}) \
    .rename(columns={'user_id': 'number_of_loyal_users'}) \
    .sort_values('number_of_loyal_users', ascending=False)
```

```python
data_pic1
```

```python
#"супер" лояльные клиенты у brand_4

pic1 = sns.barplot( x='favourite_brand', y='number_of_loyal_users', data=data_pic1)
```

```python
#чтобы определить лояльность всех клиентов, можно придумать loyalty_score

loyalty_df['loyalty_score']=loyalty_df.favourite_brand_purchases/loyalty_df.purchases
```

```python
loyalty_df
```

```python
#смотрю, как часто встречаются клиенты с разной лояльностью

pic2 = sns.distplot(loyalty_df.loyalty_score, kde=True)
```

```python
#смотрю среднюю лояльность для каждого бренда

brand_loyalty=loyalty_df.groupby('favourite_brand', as_index=False) \
    .agg({'loyalty_score': 'median', 'user_id': 'count'}) \
    .sort_values('loyalty_score', ascending = False)
```

```python
brand_loyalty
```

```python
#самые лояльные клиенты у brand_4

ax = sns.barplot( x='favourite_brand', y="loyalty_score", data=brand_loyalty)
```
